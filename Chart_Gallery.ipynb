```python
from matplotlib import pyplot as plt
import numpy as np
```

### 1. Bar chart with Error

```python
past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]
```

```python
plt.figure(figsize=(10,8))
plt.bar(range(len(years)),past_years_averages, yerr=error, capsize=10)
plt.axis([-0.5,6.5, 70,95])
ax=plt.subplot()
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)
plt.xlabel('years')
plt.ylabel('test averages')
plt.title('final exam averages')
plt.legend('year')
plt.show()
```

### 2. Bar chart 

```python
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]
```

```python
plt.figure(figsize=(10,8))
plt.axis([-0.5,5,70,90])
n = 5
int = np.arange(n)
width = 0.3
plt.bar(int, middle_school_a, width, label='middle school a')
plt.bar(int+ width, middle_school_b, width, label='middle school b')
plt.xticks(int+width/2,('Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications'))
plt.xlabel('unit')
plt.ylabel('test average')
plt.title('unit on test average')
plt.legend(loc='best')
plt.show()
```

### 3. Stack chart

```python
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

n =range(5)
width = 0.3
plt.figure(figsize=(10,8))
p1= plt.bar(n, As, width, label ='A')
p2= plt.bar(n, Bs, width, label ='B', bottom=As)
p3= plt.bar(n, Cs, width, label ='C', bottom=np.array(As)+np.array(Bs))
p4= plt.bar(n, Ds, width, label ='D', bottom=np.array(As)+np.array(Bs)+np.array(Cs))
p5= plt.bar(n, Fs, width, label ='F', bottom=np.array(As)+np.array(Bs)+np.array(Cs)+np.array(Ds))
ax=plt.subplot()
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)
plt.xlabel('unit')
plt.ylabel('Number of students')
plt.title('Grade distribution')
plt.show()
plt.savefig('my_chart.png')
```

### 6. Fill Between Chart.

```python
hours_reported =[3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25,  3.5, 3.5, 3.75, 3.75,4, 4.0, 3.75,  4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]

plt.figure(figsize=(10,8))
plt.plot(exam_scores, hours_reported, linewidth=2)
hours_lower=[i - i*0.2 for i in hours_reported]
hours_uper=[i + i*0.2 for i in hours_reported]
plt.fill_between(exam_scores, hours_lower, hours_uper, alpha=0.2)
plt.title('Time spent studying vs final exam scores')
plt.xlabel('Score')
plt.ylabel('Hours studying (self_reported)')
plt.show()
```
